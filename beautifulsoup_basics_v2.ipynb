## Ejercicio 1: Crear un objeto BeautifulSoup a partir de un string HTML básico

Crea un objeto BeautifulSoup a partir del siguiente string HTML:
```html
<html><head><title>Test</title></head><body><p>Hola Mundo!</p></body></html>
```

````python
# Solución Ejercicio 1
from bs4 import BeautifulSoup

html_doc = "<html><head><title>Test</title></head><body><p>Hola Mundo!</p></body></html>"
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup)
````

## Ejercicio 3: Buscar todos los párrafos (`<p>`) en un documento

Encuentra y muestra todos los elementos `<p>` del siguiente HTML:
```html
<html><body><p>Primer párrafo</p><p>Segundo párrafo</p></body></html>
```

````python
# Solución Ejercicio 3
html_doc = "<html><body><p>Primer párrafo</p><p>Segundo párrafo</p></body></html>"
soup = BeautifulSoup(html_doc, 'html.parser')
parrafos = soup.find_all("p") #si pongo aqui el .text, me da error porque find_all devuelve una lista de objetos, no un objeto concreto. 
for i in range(len(parrafos)):
    print(parrafos[i].text)
````

## Ejercicio 4: Buscar un elemento por su id

Dado el siguiente HTML, busca el elemento que tenga el atributo `id="unique"`:
```html
<div id="unique">Contenido único</div>
```

````python
# Solución Ejercicio 4
html_doc = "<div id='unique'>Contenido único</div>"
soup = BeautifulSoup(html_doc, 'html.parser')
elemento = soup.select_one("#unique") #find("div#unique") no funciona. select_one("unique") tampoco. Tiene que ser con el # que representa id
print(elemento.get_text())
````

## Ejercicio 5: Buscar elementos por su clase

En el siguiente HTML, encuentra todos los elementos que tengan la clase `special`:
```html
<div class="special">Uno</div><div>Dos</div><div class="special">Tres</div>
```

````python
# Solución Ejercicio 5
html_doc = "<div class='special'>Uno</div><div>Dos</div><div class='special'>Tres</div>"
soup = BeautifulSoup(html_doc, 'html.parser')
elems = soup.find_all("div", class_="special") #en los apuntes pone _class pero es class_
for i in range(len(elems)):
    print(elems[i].text)
````
